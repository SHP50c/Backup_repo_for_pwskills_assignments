```python
import pandas as pd
df4=pd.DataFrame({'Email':["sagar@g.com","har@g.com","pat@g.com","username@domain.com"]})
def user(s):
    s1=''
    for i in s:
        if i != '@':
            s1=s1+i
        else:
            break
    return s1
df4['Username']=df4['Email'].apply(user)
df4
```
